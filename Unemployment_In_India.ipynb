```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Load the dataset
file_path = "Unemployment in India.csv"
df = pd.read_csv("Unemployment in India.csv")
```

```python
df
```

```python
# Data Cleaning and Exploration
# Check for missing values
print(df.isnull().sum())

```

```python
# Explore the dataset
print(df.head())
print(df.describe())
```

```python
# Explore the dataset
print(df.head())
print(df.describe())
```

```python
# Convert 'Date' column to datetime
df[' Date'] = pd.to_datetime(df[' Date'])
```

```python
# Plot time series graph
plt.figure(figsize=(12, 6))
sns.lineplot( x=' Date', y=' Estimated Unemployment Rate (%)', data=df, hue='Region')
plt.title('Unemployment Rate Over Time by Region')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.show()
```

```python
# Regional Analysis
plt.figure(figsize=(10, 6))
sns.boxplot(x='Region', y=' Estimated Unemployment Rate (%)', data=df)
plt.title('Regional Distribution of Unemployment Rate')
plt.xlabel('Region')
plt.ylabel('Unemployment Rate (%)')
plt.show()
```

```python
# Temporal Analysis
plt.figure(figsize=(12, 6))
sns.lineplot(x=' Date', y=' Estimated Unemployment Rate (%)', data=df)
plt.title('Overall Unemployment Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.show()
```
